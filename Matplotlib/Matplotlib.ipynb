# MatplotLib Tutorial
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension is NumPy.It provides an object oriented API for embedding plots into applications using general-purpose GUI tools like Tkinter,wxPython,Qt or GTK+
Some of the major Pros of matplotlib are:
    * Generally easy to get started for simple plots
    * Support for custom labels and texts
    *Great control of every element in a figure
    *High quality output in many formats
    *Very customizable in general

```python
# compute the x and y coordinates from points on a sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine wave form')
plt.plot(x,y)
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine wave')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x,y)
```

```python
#subplot()
# compute the x and y coordinates for points on sine and cosine curves
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

# set up a subplot grid that has weight 2 and width 1,
# and set the first such subplot as active
plt.subplot(2,1,1)

#make the first plot
plt.plot(x,y_sin,'r--')
plt.title("Sine")
plt.show()

# set the second subplot as active and make the second plot
plt.subplot(2,2,2)
plt.plot(x,y_cos,'y--')
plt.title('Cosine')
plt.show()
```

```python
## Bar plot

x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

# Histograms

```python
a=np.array([22,87,5,43,56,73,55,11,20,51,5,79,31,27])
plt.hist(a)
plt.title('histogram')
plt.show()
```

```python
a=np.array([22,87,5,43,56,73,55,11,20,51,5,79,31,27])
plt.hist(a,color='y',bins=15)
plt.show()
```

# Box Plot using Matplotlib

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True);
```

# Pie Chart

```python
# Data to plot
labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0)
# plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
